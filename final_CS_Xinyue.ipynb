# Binary Search

## Question ($\star\star$, 30 marks)
- Problem: 
    - follow up on Exercise 3.15 in Chapter 3 of lecture note
    - find a target in a $m \times n$ matrix
    - here we can assume that 
        - each row in the matrix is sorted in ascending order
        - each column in the matrix is sorted in ascending order
        - $m < n$
    - see examples in the testing cases
    - find the solution with the complexity below
- Complexity: 
    - $O(m + log(n)$ time
    - $O(1)$ space
- **Logic: # Implement me (5 marks)**
First find the possible row that contains the target. Then use binary search in that row to find the target. 

```python
# Implementation
def fun_21(matrix, target):
    """
    Find a target in a m * n matrix
    
    Parameters
    ----------
    matrix : list[list[int]] (i.e., a list of list of integers)
    target : an integer
    
    Returns
    ----------
    True  : if the target is in the matrix
    False : otherwise
    """
    
    # Implement me (25 marks)
    row = len(matrix)
    
    if row == 0:
        return False
    
    col = len(matrix[0])
    
    if col == 0:
        return False
    
    for i in range(row):
        left = 0
        right = col-1
        
        # Find the possible row
        if matrix[i][left] <= target and matrix[i][right] >= target:
            # Use Binary Search in the row
            while left <= right:
                mid = left + (right - left) // 2
                if matrix[i][mid] == target:
                    return True
                elif matrix[i][mid] > target:
                    right = mid - 1
                elif matrix[i][mid] < target:
                    left = mid + 1

    return False    
```

## Question ($\star\star$, 20 marks)
- Problem: 
    - follow up on Exercise 5.5 in Chapter 5 of lecture note
    - find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    - here, the path starts from the root and ends at a leaf node
    - see examples in the testing cases
    - find the solution with the complexity below    
    - you must use DFS to design the solution
- Complexity: 
    - $O(n)$ time
    - $O(n)$ space
    - where $n$ is the number of nodes in the tree
- **Logic: # Implement me (5 marks)**
If root is None, then there is no path. Either the left subtree or the right subtree has a correct sum is OK. When go to the next level, the target should minus the value of the root first. If the value of a leaf is equal to the current target, then we find the path.

```python
# Implementation
def fun_31(root, target):
    """
    Find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    
    Parameters
    ----------
    root : the root of a binary tree
    target : an integer
    
    Returns
    ----------
    True,  if there is such a path
    False, otherwise
    """
    
    # Implement me (15 marks)
    # Base
    if root is None:
        return False
    
    if root.left is None and root.left is None:
        # This is end of the path
        if root.val == target: # find the path
            return True
    
    return fun_31(root.left, target - root.val) or fun_31(root.right, target - root.val) 
    # either the left subtree or the right subtree has a correct sum is OK so use 'or'
    # new target should be the old target minus the value of the root
```

## Question ($\star$, 15 marks)
- Problem: 
    - follow up on Exercise 5.10 in Chapter 5 of lecture note
    - find the number of (vertically or horizontally) connected deserts in a map
    - unlike Exercise 5.10 where the map after running the solution does not have to be the same as the input one, here the map must be the same as the input
    - that is, you must reverse any change made in the map after DFS <font color='red'>(you may consider this as a hint)</font>
    - see examples in the testing cases
    - find the solution with the complexity below   
    - note that you cannot make a copy of the map since that requires $O(n)$ space (where $n$ is the number of entries in the map)
- Complexity: 
    - $O(n)$ time
    - $O(1)$ space
- **Logic: # Implement me (5 marks)**
If an entry in the map is '1', then in the helper function, we change all its connected '1' into '2', and increase the count of islands by 1.
After the DFS, we change all the '2' into '1' to restore the map.

```python
# Reference: Part of the code below is from chapter_5_tree_search_solution.ipynb
# Implementation
def fun_32(matrix):
    """
    Find the number of (vertically or horizontally) connected deserts in a map (represented by a matrix)
    
    Parameters
    ----------
    matrix : a matrix where 1 denotes desert and 0 denotes grassland
    
    Returns
    ----------
    the number of connected deserts
    """

    # Corner case
    if len(matrix) == 0 or len(matrix[0]) == 0:
        return 0

    m, n = len(matrix), len(matrix[0])
    # The number of connected deserts
    count = 0

    for i in range(m):
        for j in range(n):
            if matrix[i][j] == 1:
                count += 1
                # DFS: remove '1's conntected to entry [i, j]
                helper(matrix, m, n, i, j)
                
    # Implement me (5 marks)
    # Restore the matrix
    # Change 2 to 1
    for i in range(m):
        for j in range(n):
            if matrix[i][j] == 2:
                matrix[i][j] = 1
                
    return count
```

```python
def helper(matrix, m, n, i, j):
    """
    DFS: remove '1's conntected to entry [i, j]

    Parameters
    ----------
    matrix : a matrix where 1 denotes desert and 0 denotes grass land
    m : the nubmer of rows in the matrix
    n : the nubmer of columns in the matrix
    i, j : entry [i, j]
    """

    # Implement me (5 marks)
    # If either i or j is not valid, or matrix[i][j] is not 1
    if i < 0 or i >= m or j < 0 or j >= n or matrix[i][j] != 1:
        return

    # Change 1 to 2.
    # 2 means it was 1 originally.
    matrix[i][j] = 2

    # Remove '1' above [i, j]
    helper(matrix, m, n, i - 1, j)
    # Remove '1' below [i, j]
    helper(matrix, m, n, i + 1, j)
    # Remove '1' on the left-hand side of [i, j]
    helper(matrix, m, n, i, j - 1)
    # Remove '1' on the right-hand side of [i, j]
    helper(matrix, m, n, i, j + 1)
```

# Breadth First Search (35 marks)

## Question ($\star\star$, 20 marks)
- Problem: 
    - follow up on Question 3.1 above
    - find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    - see examples in the testing cases
    - find the solution with the complexity below
    - you must use BFS to design the solution
    - the node value can be modified <font color='red'>(you may consider this as a hint)</font>
- Complexity: 
    - $O(n)$ time
    - $O(1)$ space
- **Logic: # Implement me (5 marks)**
Change the node value to (target - all the parent nodes' values). If a leaf's value is 0, then we find the path.

```python
# Implementation
from collections import deque

def fun_41(root, target):
    """
    Find whether there is a path in the binary tree such that the sum of value of nodes in the path equals a target
    
    Parameters
    ----------
    root : the root of a binary tree
    target : an integer
    
    Returns
    ----------
    True,  if there is such a path
    False, otherwise
    """
    
    # Implement me (15 marks)
    if root is None:
        return False
    
    # Minus root value from the target
    root.val = target - root.val    
    # Initialize the queue
    queue = deque([root])
    
    # Level order traversal
    while len(queue) > 0:
        # Get the length of the queue
        n = len(queue)    
        for _ in range(n):
            # Remove the first node in the queue
            node = queue.popleft()
            if node is None:
                continue
            # Minus the leaves's value from the target
            if node.left is not None:
                node.left.val = node.val - node.left.val
                queue.append(node.left)
            if node.right is not None:
                node.right.val = node.val - node.right.val
                queue.append(node.right)
            # If the node is a leaf
            if node.left is None and node.right is None:
                #print(node.val)
                if node.val == 0:
                    return True
                
    return False
```

## Question ($\star$) (15 marks)
- Problem:
    - follow up on Exercise 6.10 in Chapter 5 of lecture note
    - unlike Exercise 6.10 where we should find the bottom-left node in a binary tree, here we should find the bottom-right node
    - see examples in the test cases
    - find the solution with the complexity below
    - you must use BFS to design the solution
- Complexity:
    - $O(n)$ time
    - $O(n)$ space
- **Logic: # Implement me (5 marks)**
Record the right most node's value on each level. Return the last right most value.

```python
# Implementation
def fun_42(root):
    """
    Find the bottom-right node in a binary tree
    
    Parameters
    ----------
    root : the root of a tree
    
    Returns
    ----------
    The value of the bottom-right node in a binary tree
    """

    # Implement me (10 marks)
    # Corner case
    if root == None:
        return root
    
    bottom_right = root.val
    
    # Initialize the queue
    queue = deque([root])
    
    while len(queue) > 0:
        # Get the length of the queue
        n = len(queue)
        for i in range(n):
            # Remove the first node in the queue
            node = queue.popleft()
            
            # Update bottom_right using the value of the right most node on each level
            if i == n-1:
                bottom_right = node.val

            if node.left is not None:
                # Add the value of the left child to the end of the queue
                queue.append(node.left)
            if node.right is not None:
                # Add the value of the right child to the end of the queue
                queue.append(node.right)

    return bottom_right
 
```
